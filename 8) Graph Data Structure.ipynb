#### Before proceeding with our main code, a basic reminder between list append and adding.
- mylist.append(x) -> It means add x into the current list
- mylist = mylist+x -> it means create a brand new list containing all old elements plus x

```python
# difference b/w path.append(start) and path = path +[start]
path = ["Mumbai", "Paris"]
path.append("Dubai")     # added "Dubai" directly into the existing list.
print(path)
```

```python
path = ["Mumbai", "Paris"]
path = path + ["Dubai"]   # Python creates a new list and assigned it back to path.
print(path)
```

```python
path = ["Mumbai", "Paris"]
print(id(path))
path.append("Dubai")
print(id(path))

# You can see, if we are using the append method the variable/object id is same. In short, python modified the existing list.
```

```python
path = ["Mumbai", "Paris"]
print(id(path))
path = path + ["Dubai"]
print(id(path))

# Different ID.
# Python created a new list and assigned it back to path
```

```python
a = [1, 2]
b = a
b.append(3)
print(a)
print(b)
# Here, both a and b variable, point to the same list. Even, if I append something in b, it will also modift a
```

```python
a = [1,2]
b = a
b = b + [3]
print(a)
print(b)

# Here, it will not modify the variable a, if b is changed. As both are different objects now.
```

 ### Implementing Graph in Python

<img src=".\Reference Images\Route.png" width="300">


```python
class Graph:
    def __init__(self, edges):
        self.edges = edges   # Store original list of edges/routes
        # Dictionary that will store graph in adjacency-list format. It is to represent direct connection between two nodes.
        # Example:
        # {
        #   'Mumbai': ['Paris', 'Dubai'],
        #   'Paris': ['Dubai', 'New York']
        # }
        self.graph_dict = {}
        for start, end in edges:  # Build adjacency list from edge list
            # If source city already exists in dictionary, add destination city to existing list
            if start in self.graph_dict:
                self.graph_dict[start].append(end)
            else:   # # Otherwise create a new entry
                self.graph_dict[start] = [end]
        print("Graph Dict:", self.graph_dict)   # print the adjacency list

    # -------------------------------------------------------
    # Find ALL possible paths from start to end
    # -------------------------------------------------------
    def get_paths(self, start, end, path=[]):
        # Create a new path list and add current city.
        # We create a new list to avoid modifying the same path object in recursive calls.
        # If we use path.append(start), it will modify the same list object in recursive calls. It will not show multiple objects.
        path = path + [start]    #It creates a new list everytime.

        # Base Case: If destination reached, return the complete path
        if start == end:
            return [path]   #path[], means list of complete path, not single path like return path like return [['Mumbai','Paris','New York']]

        # If current city has no outgoing route,no path exists from here. It will check from adjacent list if key is there for the current city. No key means no available path from here to any location.
        if start not in self.graph_dict:
            return []

        paths = []   # Store all paths discovered from current city

        # Explore every neighboring city
        for node in self.graph_dict[start]:
            # Prevent circular traversal. It will check in the open object of current path if particular node is already there or not. In start path = ['Mumbai']. So if mumbai is repeated anywhere in node again, it will not consider it.
            # Example: Mumbai -> Paris -> Mumbai -> Paris ...
            if node not in path:   # Run code only when condition is true. Otherwise, skip that block. e.g. if mumbai or any location is repeated skip it.
                # Ask neighboring city to find paths
                new_paths = self.get_paths(node, end, path)  # E.g.  self.get_paths("Paris","New York",["Mumbai"])
                # Add returned paths into final list
                for p in new_paths:
                    paths.append(p)

        # Return all discovered paths
        return paths

    # -------------------------------------------------------
    # Find SHORTEST path from start to end
    # -------------------------------------------------------
    def get_shortest_path(self, start, end, path=[]):
        path = path + [start]   # Add current city to path

        # Base Case:Destination reached
        if start == end:
            return path

        # No route available from current city
        if start not in self.graph_dict:
            return None

        shortest_path = None    # Variable to store shortest path found so far
        for node in self.graph_dict[start]:          # Explore every neighboring city
            if node not in path:    # Avoid revisiting cities already in path
                # Ask neighboring city for shortest path
                sp = self.get_shortest_path(node, end, path)
                # If a valid path exists
                if sp:
                    # First path found OR shorter path found
                    if shortest_path is None or len(sp) < len(shortest_path):
                        shortest_path = sp

        return shortest_path   # Return shortest path found

```

```python
if __name__ == '__main__':

    routes = [
        ("Mumbai", "Paris"), # Every tuple indicates one route
        ("Mumbai", "Dubai"),
        ("Paris", "Dubai"),
        ("Paris", "New York"),
        ("Dubai", "New York"),
        ("New York", "Toronto"),
    ]

    route_graph = Graph(routes)

    start = "Mumbai"
    end = "New York"

    print(f"All paths between: {start} and {end}: ",route_graph.get_paths(start,end))
    print(f"Shortest path between {start} and {end}: ", route_graph.get_shortest_path(start,end))

    start = "Dubai"
    end = "New York"

    print(f"All paths between: {start} and {end}: ",route_graph.get_paths(start,end))
    print(f"Shortest path between {start} and {end}: ", route_graph.get_shortest_path(start,end))
```
