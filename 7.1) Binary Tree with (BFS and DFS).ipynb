```python
# Creating program for binary search tree node class. (Each node can have at most 2 children, and it should be in specific order. Also no duplicates)

class BinarySearchTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

    def add_child(self,data):
        if data == self.data:   # prevent duplicate entry.
            return
        if data < self.data:   # add data in left subtree
            if self.left:      # check whether left child exists. Same as (if self.left is not None:)
                self.left.add_child(data)  # Call add_child function again from left node to add the data below the left child or on level 2.(Recursive function)
            else:    # If condition false, add data in left side
                self.left = BinarySearchTreeNode(data)
        else:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = BinarySearchTreeNode(data)

    # below function will return a list of elements from left,root, right.
    def in_order_traversal(self): #(left -> Root -> Right)
        elements = []
        # If I have a left child, give me all your nodes in a sorted manner.
        if self.left:    #visit left node
            elements += self.left.in_order_traversal()

        # visit base node (add myself)
        elements.append(self.data)

        # # If I have a right child, give me all your nodes in a sorted manner.
        if self.right:
            elements += self.right.in_order_traversal()

        return elements

    def search(self,val): #search a value in tree
        if self.data == val:
            return True
        if val < self.data:  #check if value in left subtree
            if self.left:
                return self.left.search(val)
            else:
                return False
        if val > self.data:
            if self.right:
                return self.right.search(val)
            else:
                return False

    def find_max(self):
        if self.right is None: # if no value on right, main node is max
            return self.data
        return self.right.find_max()   #recursively call the function again from next node to until upper statement is true

    def find_min(self):
        if self.left is None: # if no value on right, main node is max
            return self.data
        return self.left.find_min()


    def delete(self,val):
        # Step 1: Search for the value to be deleted.
        # If value is smaller than current node, it can only exist in the left subtree.
        if val < self.data:
            if self.left: # # Check whether left subtree exists.
                # Recursively delete from left subtree. And the returned node becomes the new left child.
                self.left = self.left.delete(val)

        # If value is greater than current node,it can only exist in the right subtree.
        elif val > self.data:
            if self.right:         # Check whether right subtree exists.
                # Recursively delete from right subtree. and the returned node becomes the new right child.
                self.right = self.right.delete(val)

        # If neither of the above condition is true, then val == self.data. And then we have found the node to delete.
        else:
            # --------------------------------------------------
            # CASE 1: Node has no children (Leaf Node)
            # Example:
            #     10
            # Simply remove it by returning None.
            # --------------------------------------------------
            if self.left is None and self.right is None:
                return None
            # --------------------------------------------------
            # CASE 2: Node has only right child
            # Example:
            #    10
            #      \
            #       15
            # Replace current node with its right child.
            # --------------------------------------------------
            if self.left is None:
                return self.right
            # --------------------------------------------------
            # CASE 3: Node has only left child
            # Example:
            #       10
            #      /
            #     5
            # Replace current node with its left child.
            # --------------------------------------------------
            if self.right is None:
                return self.left

            # --------------------------------------------------
            # CASE 4: Node has two children
            #
            # Example:
            #         15
            #        /  \
            #      10    20
            #           /
            #         18
            #
            # We cannot directly delete 15 because it
            # would break the BST structure.
            #
            # Solution:
            # 1. Find smallest value in right subtree.
            # 2. Replace current node's value with it.
            # 3. Delete the duplicate value from right subtree.
            # -------------------------------------------------
            min_val = self.right.find_min()           # Find the smallest value in right subtree.
            self.data = min_val                      # Replace current node's data with successor value.
            self.right = self.right.delete(min_val)  # Delete duplicate node from right subtree.
            # Alternate approach for above 3 lines, to use max value from left node and use it to replace the current node's data
            # max_value = self.left.find_max()
            # self.data = max_value
            # self.left = self.left.delete(max_value)

        return self    # Return current node (possibly modified) so that parent can reconnect to it.

    def calc_sum(self):    # Sum of all elements in the tree. Using inorder traversal
        print("Sum of all elements in tree is: ",sum(self.in_order_traversal()))

    def post_order_traversal(self):  # post order traversal (left -> Right -> Root)
        elements = []
        if self.left:   # if I have a left child, give me nodes from left,right and root
            elements += self.left.post_order_traversal()
        if self.right:
            elements += self.right.post_order_traversal()
        elements.append(self.data)
        return elements

    def pre_order_traversal(self): #(Root -> left -> Right)
        elements = [self.data]
        if self.left:
            elements += self.left.pre_order_traversal()
        if self.right:
            elements += self.right.pre_order_traversal()
        return elements
```

```python
#Helper method to build tree. It will take elements as input and build the tree for us.
def build_tree(elements):
    root = BinarySearchTreeNode(elements[0])

    for i in range (1,len(elements)):
        root.add_child(elements[i])
    return root

if __name__ == '__main__':
    numbers = [17,4,1,20,9,23,18,24]
    numbers_tree = build_tree(numbers)
    # print(numbers_tree.in_order_traversal())
    # numbers_tree.delete(17)
    # print(numbers_tree.data)     # return root node or first value
    # print(numbers_tree.in_order_traversal())
    # numbers_tree.calc_sum()
    print(numbers_tree.post_order_traversal())
    print(numbers_tree.pre_order_traversal())
```
