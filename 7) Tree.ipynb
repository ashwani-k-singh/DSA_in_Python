### Tree:
A Tree is a non-linear hierarchical data structure made up of nodes connected by edges.

Unlike arrays, linked lists, stacks, and queues (which are linear), a tree represents data in a parent-child relationship.

<img src=".\Reference Images\tree.png" width="900">


```python
# Python don't have a built-in tree data structure. So we will be creating a class for it.
class TreeNode:   #class which defines how a tree node look like
    def __init__(self, data):  # each node contain data, children and parent
        self.data = data
        self.children = []   #we are overwriting it, no need to define above
        self.parent = None

# Alternate Method
# def __init__(self, data, children=None, parent=None):
#    self.data = data
#    self.children = children if children else []
#    self.parent = parent

    def get_level(self): # Find how deep a node is in the tree.
        level = 0        #root level which is 0
        p = self.parent
        while p:          #if self.parent is not None run loop
            level += 1
            p = p.parent  # now move one more level up

        return level   # for root, it will return 0 since self.parent is None

    def print_tree(self):   # hierarchy print
        spaces = ' ' * self.get_level() * 3  #create spaces for printing based on node level
        #add |__ after spaces while printing for parent is not None
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + self.data)
        if self.children: # (If node has children: visit every child.)
            for child in self.children:
                child.print_tree()  #calls the same function again (recursion)

    def add_child(self, child):  # Connect a child node to a parent node.
        # Suppose laptop.add_child(TreeNode("Mac"))
        child.parent = self   # means mac.parent=Laptop
        self.children.append(child) # Laptop.children.append(Mac)- add in list children

```

```python
def build_product_tree():
    root = TreeNode("Electronics")

    laptop = TreeNode("Laptop")
    laptop.add_child(TreeNode("Mac"))
    laptop.add_child(TreeNode("Surface"))
    laptop.add_child(TreeNode("Thinkpad"))

    cellphone = TreeNode("Cell Phone")
    cellphone.add_child(TreeNode("iPhone"))
    cellphone.add_child(TreeNode("Google Pixel"))
    cellphone.add_child(TreeNode("Vivo"))

    tv = TreeNode("TV")
    tv.add_child(TreeNode("Samsung"))
    tv.add_child(TreeNode("LG"))

    root.add_child(laptop)
    root.add_child(cellphone)
    root.add_child(tv)

    root.print_tree()

if __name__ == '__main__':
    build_product_tree()
```
