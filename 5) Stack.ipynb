### STACK
A Stack is a linear data structure that follows: LIFO Principle- Last In First Out.

Meaning: The element inserted last is removed first.

<img src=".\Reference Images\stack.png" width="700">


#### Why should we use stack?
Stacks are used when:
- latest data must be processed first
- reversing order is needed
- tracking function calls
- undo/backtracking operations

##### Some of the Stack Use cases:
1. Browser Back Button	- Every visited page is pushed onto a stack. Clicking Back pops the latest page and returns to the previous one.
2. Undo Operation (Ctrl + Z) -	Text editors like Word, VS Code, and Photoshop store actions in a stack. The most recent action is undone first.
3. Function Calls (Call Stack) -	When a function is called, it is pushed onto the call stack. After execution, it is popped off. This is how recursion works internally.
##### Note: A stack can be implemented using: list/ linked list/  collections.deque/ queue.LifoQueue





#### Stack vs List
A stack is a data structure behavior, while a Python list is a general container.  List can do append, insert, pop from anywhere.
However, stack operation is restricted to top insertion and top deletion.

#### 2) Stack Using Linked List

#### Implement Stack Class Using deque

```python
from collections import deque

class Stack:    #create customer stack data structure
    def __init__(self):     #constructor, it run automatically when  s=stack()
        self.container = deque()  #deque is Double Ended Queue which is a built-in Python data structure optimized for fast insertion and deletion. Think of it like: deque([]) an empty container.

# After: self.container = deque()
# memory conceptually becomes: container → []

    def push(self,val):      #Add element to top of stack. Also Why append at end? - Because stack top is maintained at the right side: Last inserted item becomes top.
        self.container.append(val)

    def pop(self):   # Remove top item LIFO
        return self.container.pop()

    def peek(self):   # View top item without removing it
        return  self.container[-1]

    def is_empty(self):  # Check whether stack is empty
        return len(self.container)==0

    def size(self):   # Return number of elements
        return len(self.container)
```

```python
s = Stack()
s.push(5)
print(s.is_empty())
print(s.pop())
```
