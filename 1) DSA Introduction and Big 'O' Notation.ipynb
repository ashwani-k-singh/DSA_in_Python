### DSA Introduction:
Data Structures are building blocks or raw material for any software programs. They are the containers storing data in a specific memory layout.

DSA is the study of how to store data efficiently and solve problems efficiently.

#### Data Structure = How data is stored

Examples:

- Array → Store items in a list
- Linked List → Store items connected by links
- Stack → Last In, First Out (like a stack of plates)
- Queue → First In, First Out (like a line at a ticket counter)
- Tree → Store data in a hierarchical structure

#### Algorithm = Steps to solve a problem

Examples:

- Searching for a number in a list
- Sorting numbers in ascending order
- Finding the shortest path between two places


## Big 'O' Notation
Big O notation is a way to describe how fast or slow an algorithm grows as input size increases. It measures:

- time complexity (speed)
- space complexity (memory usage)

Big O does not measure exact seconds. Instead, it measures growth behavior. It matters since for small programs everything seems fast, but for large programs, inefficient programs becomes unusable.

<img src=".\Reference Images\Big O notation.png" width="500">

Big O ignores:
- constants
- small terms
- e.g. 3n² + 5n + 10, then Big O becomes O(n²) because largest-growing term dominates.  It even ignores constant 3 as it is not major focus.

```python
# O(1) - Constant Time
# Here Execution time stays same regardless of input size. e.g. accessing first element of an array is always instant.Whether array has:3 elements or 3 million elements

arr = [10,20,30]
print(arr[0])
```

```python
# O(n) - Linear Time
# Time grows directly with input size. E.g. If array length doubles: work doubles.

for i in arr:      # print all elements in the array
    print(i)
```

```python
# O(n²) — Quadratic Time
# E.g. Nested loops. It is very expensive and not good for scaling.

for i in arr:                # Outer loop runs n times
    for j in arr:            # For each i, inner loop runs n times
        print(i, j)          # Total executions = n × n = n²

# If n=10, then 100 operation.       If n=100, 10000 operations
```

```python
# O(log n) — Logarithmic Time
# The problem size is reduced by half in every iteration. This is much faster than checking every element one by one.
# Example:  1000 items  → ~10 checks     ////  1,000,000 items → ~20 checks

# Best example - Binary Search(works on sorted array only).
# No need to focus too much on below code, we will study it later. It is for reference here only.

def binary_search(ary, target):
    # Start searching from the first element
    left = 0
    # End searching at the last element
    right = len(ary) - 1
    # Continue until the search range becomes invalid
    while left <= right:
        # Find the middle index// = Floor Division (returns an integer)
        mid = (left + right) // 2
        # Target found at middle position
        if ary[mid] == target:
            return mid
        # Target is larger than middle element, then ignore the left half and search the right half
        elif ary[mid] < target:
            left = mid + 1
        # Target is smaller than middle element,then Ignore the right half and search the left half
        else:
            right = mid - 1
    # Executed when target is not present in the array, -1 is commonly used to indicate "Not Found"
    return -1


# Sorted array (required for Binary Search)
numbers = [2, 4, 6, 8, 10, 12, 14]

# Search for 10
result = binary_search(numbers, 10)

print("Index:", result)

# Binary Search is logarithmic because each comparison eliminates half of the remaining elements, causing the search space to shrink exponentially.
```

Same is valid for space complexity.

Suppose input size becomes bigger:
- So does your program: use same memory?
- or need much more memory?


```python
# O(1) Space


def add(a, b):
    return a + b      # This uses almost fixed memory: only few variables

print(add(10, 20))
# Even if numbers become huge: memory usage stays almost same. Also, no extra variable created to store result.
```

```python
# O(n²) Space Complexity
# Memory usage grows quadratically because we create an n × n matrix (n rows and n columns).
# Example:
# n = 2 → 2 × 2 = 4 elements
# n = 3 → 3 × 3 = 9 elements
# n = 100 → 100 × 100 = 10,000 elements

matrix = []                    # Empty list to store all rows
n = 2                          # Matrix size (2 rows × 2 columns)
for i in range(n):             # Outer loop runs n times (creates rows)
    row = []                   # Create a new empty row
    for j in range(n):         # Inner loop runs n times (creates columns)
        row.append(0)          # Add a value to the current row
    matrix.append(row)         # Add completed row to matrix

print(matrix)
```

O(n²) Space means the program stores a two-dimensional structure whose size grows as n × n. A matrix is one of the most common examples of quadratic space complexity.
