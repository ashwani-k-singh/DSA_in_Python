### 3) Linked List
A Linked List is a linear data structure where elements are stored in nodes. Each node contains:

1. Data – the actual value.
2. Next – a reference (link) to the next node.

Unlike Python lists, linked list elements are not stored in contiguous memory locations.

#### Why do we need linked list?
In python list, if we need to insert a value at the beginning then python must shift all elements one position to the right. It results in time complexity O(n).

A linked list avoid this shifting by changing the links between nodes. It makes insertion easier and we don't need to pre allocate space.

```python
# CREATING SINGLY LINKED LIST

class Node:          # class node which defines what a node looks like
    def __init__(self,data=None, next=None):     #constructor
        self.data = data    # data member which stores actual value. By default, it is None
        self.next = next    # next member that stores reference/pointer to the next node

class LinkedList:     #Linked list class which manages all nodes
    def __init__(self):
        self.head = None   #points to head or first element of linked list. Initially head → None which means no nodes exist

    def insert_at_beginning(self,data):    #insert node before current head -> O(1)
        node = Node(data,self.head)    # here self.head() refers to next. It stores pointer of next node. Initially, it is none for first input.
        self.head = node              #self.head changes the head node to updated head which we can use to point next time

# What does head store? -> self.head stores a reference to the first node only:
# But Because every node knows the next node: node.next
# Python can reach all nodes by following the chain.
# Think of it like train coaches:
# head
#  ↓
# [20] -> [10] -> [70] -> None
# You only need the first coach. Every coach tells you where the next coach is.This is how all the inputs are stored which we can use to print or call later.

    def print(self):              # Traverse entire linked list and display it
        if self.head == None:
            print("Linked List is empty")
            return

        itr = self.head   # itr we will use as iterator or temporary pointer here.
        llstr = ''        # String variable which is used to build output.
        while itr:        # Same as while itr != None: continue till end
            llstr += str(itr.data) + ' --> '   # first iteration itr.data which is self.head. So it becomes llstr += "10 --> "
            itr = itr.next      # move to next node. When reaches None, loop ends
        print(llstr)

    def insert_at_end(self,data):     # add  node at tail -> O(n)
        if self.head is None:         # same as if self.head == None:
            self.head = Node(data,None)  # Create first node
            return
        itr = self.head          # Iterator which will be executed if self.head is not None.
        while itr.next:         # Move until the last node. It will run until itr.next is None
            itr = itr.next

        itr.next = Node(data,None) # attach new node, itr.next = Node(data,None)

    def insert_values(self, data_list):  # Insert multiple values.
        self.head = None               # Resets initial list or deletes access to old list whenever we call insert_values
        for data in data_list:         #insert data at end one by one
            self.insert_at_end(data)

# The old linked list is still present in memory for a short time, but our program no longer has a reference to it.
# Python's Garbage Collector eventually notices: "These Node objects are unreachable; no variable references them." and frees their memory automatically. We don't control exactly when this happens, but it will happen eventually.

    def get_length(self):        # Counts nodes
        count = 0                #initialize count =0
        itr = self.head          # Iterator which will refers to self.head
        while itr:
            count += 1
            itr = itr.next
        return count

    def remove_at(self,index):   # Deletes node at a given index in linked list
        if index<0 or index>=self.get_length(): # Protects against invalid indexes. e.g. remove at(-1) or index more than list length
            raise Exception("Invalid index")

# Below self.head = self.head.next is a Special case - removing first node.
# Normally removal works by previous.next, however for first node, there is no previous node. So, if we remove the first node directly, there will be no access to next nodes.
# What does "remove_at(2)" mean? - Many beginners think: We physically remove the node from memory.That's not how linked lists work. In a linked list, deletion means: Break the connection to the node. Think like roads. Imagine cities: Delhi → Mumbai → Pune → Goa
# Roads:Delhi road -> Mumbai, Mumbai road -> Pune,Pune road -> Goa.
# Suppose Pune disappears. So, Do we need to destroy Pune first? No.
# We simply build a new road: from Mumbai to Goa and now nobody travels through pune which becomes isolated now.
# So we will just isolate the node we want to remove and bypass it using the previous node.

        if index==0:
            self.head = self.head.next     # index 0 becomes isolated and index 1 becomes index 0
            return

        count = 0
        itr = self.head    #iterator
        while itr:         # run loop till itr is None
            if count == index-1:   # Move until count == index -1 (previous node before remove index)
                itr.next = itr.next.next   # now the itr.next becomes itr.next.next. In short the next bypasses one node.
                break
            itr = itr.next
            count += 1

    def insert_at(self,index,data):              # insert at any position.
        if index<0 or index>self.get_length():   # protect against invalid indexes.
            raise Exception("Invalid index")
        if index==0:                            # index 0 means to insert at beginning
            self.insert_at_beginning(data)
            return

        count = 0      # Used to keep track of which node we're currently visiting.
        itr = self.head   # temporary pointer used for traversal.
        while itr:
            if count == index-1:   # -1, since we need to change pointer of previous node of the location at which we wish to insert.
                node = Node(data,itr.next)     # here itr becomes one node before and itr.next its next node. So the new inserted node will create two separate paths.
                itr.next = node                # merge both paths. Both separate chains created after insertion are merged.
                break
            itr = itr.next
            count += 1

    def insert_after_value(self,data_after,data_to_insert):
        itr = self.head
        while itr:
            if itr.data == data_after:
                node = Node(data_to_insert,itr.next)
                itr.next = node
                break
            itr = itr.next
        else:               # else executes only if the loop finishes normally without hitting break.
            raise Exception("Invalid input, no data entry ",data_after,"exist in linked list")     # it will throw error. We can use try, except method to normally print the exception remark.

    def remove_by_value(self,data):
        itr = self.head

        if itr.data == data:
            self.head = self.head.next     # index 0 becomes isolated and index 1 becomes index 0
            return

        while itr and itr.next:
            if itr.next.data == data:   #this will throw error if itr.next.data is None while comparing. In situation where the data value not in LL.
                itr.next = itr.next.next
                break
            itr = itr.next
        else:
            raise Exception(f"Invalid input, no data entry '{data}' exist in linked list")
```

```python
ll = LinkedList()
ll.insert_at_beginning(10)
ll.insert_at_end(70)
ll.print()
```

```python
ll.insert_values(["banana","mango"])
ll.print()
print("length:",ll.get_length())
```

```python
ll.remove_at(1)  #Remove node at index 1
ll.print()
```

```python
ll.insert_at(0,"papaya")
ll.insert_at(1,"jackfruit")
ll.print()
```

```python
print(ll.head.data)  # using python built-in print function to access data of head.
```

```python
ll.insert_after_value('banana','orange')   #insert after banana
ll.print()
```

```python
ll.remove_by_value('jack')  #Remove jack which is not in list. It will throw exception: Invalid input, no data entry 'jack' exist in linked list
```

```python
# Double Linked List Program
class ll_node:          # class node which defines what a node looks like
    def __init__(self,data=None, next=None, prev = None):     #constructor
        self.data = data    # data member which stores actual value. By default, it is None
        self.next = next    # next member that stores reference/pointer to the next node
        self.prev = prev   # prev member that stores pointer to the previous node


class Doubly_ll:            # Doubly linked list class with both previous and next reference node
    def __init__(self):   #constructor
        self.head = None    # Gives instant access to first node
        self.tail = None    # Gives instant access to last node, making backward printing much easier


    def insert_at_start(self,data):   # insert node before the current head
        if self.head is None:         #check for empty list
            node = ll_node(data)
            self.head = node
            self.tail = node         #tail is required when list is empty as we can directly access this element if anything is added in starting
            return
        # Now for non-empty list
        node = ll_node(data, self.head,None)   #previous is none always here
        self.head.prev = node     # old head now point to new head
        self.head = node   # Now move head to new head


    def print_forward(self):              # Traverse entire linked list and display it
        if self.head == None:
            print("Linked List is empty")
            return

        itr = self.head   # itr we will use as iterator or temporary pointer here.
        llstr = ''        # String variable which is used to build output.
        while itr:        # Same as while itr != None: continue till end
            llstr += str(itr.data) + ' --> '   # first iteration itr.data which is self.head. So it becomes llstr += "10 --> "
            itr = itr.next      # move to next node. When reaches None, loop ends
        print(llstr)

    def print_backward(self):
        if self.tail is None:
            print("Linked List is empty")
            return
        itr = self.tail
        llstr = ''
        while itr:
            llstr += str(itr.data) + ' <-- '
            itr = itr.prev
        print(llstr)

    def insert_at_end(self,data):
        if self.head is None:          # If list is empty create new head which is also a tail for later references
            node = ll_node(data)
            self.head = node
            self.tail = node
            return                      # close function, when if condition is TRUE or executed

        node = ll_node(data,None,self.tail)  #It will point to self.tail and no next as it is None (we are inserting at end)
        self.tail.next = node         #now new node will become last node
        self.tail = node              #now the tail will become or updated to last node

    def insert_values(self, data_list):  # Insert multiple values at end of ll. (Note- we will not reset previous list here)
        for data in data_list:         #insert data at end one by one
            self.insert_at_end(data)

    def get_length(self):        # Counts nodes
        count = 0                #initialize count =0
        itr = self.head          # Iterator which will refers to self.head
        while itr:
            count += 1
            itr = itr.next
        return count

    def remove_at(self,index):   # Deletes node at a given index in linked list
        if self.head is None:
            print("Linked List is empty")
            return
        length = self.get_length()  # to reduce time complexity we will calculate length here. And we can use legth variable further

        if index<0 or index>=length:   #if index is less than 0 or equal/greater than list length - print invalid index
            raise Exception("Invalid index")      # Once it is executed, it will jump out of the function directly. So, no need to use the break keyword here.

        if index == 0 and length!=1:   #without self.get_length()!=1, next line will throw error when list length is 1.
            self.head.next.prev = None     # Need to remove the access of first node from 2nd node
            self.head = self.head.next     # value at 2nd position becomes head
            return

        if index == 0 and length ==1:
            self.head = None     # Empties the list
            self.tail = None
            return

        if index == length - 1:        # Means if we wish to remove the tail or last value of the linked list
            self.tail.prev.next = None       # remove access of last value from 2nd last value
            self.tail = self.tail.prev       # Make the 2nd last value - last value
            return

        # For all remaining conditions or to remove something in between the list
        itr = self.head
        count = 0
        while itr:  #itr is 20
            if index == count:                      #10 20 30 40   -> remove 20
                itr.next.prev = itr.prev
                itr.prev.next = itr.next
                return
            count += 1
            itr = itr.next

    def insert_at(self,index,data):   # Insert node at a specific index
        length = self.get_length()
        if index<0 or index>length:   #if index is less than 0 or equal/greater than list length - print invalid index
            raise Exception("Invalid index")      # Once it is executed, it will jump out of the function directly. So, no need to use the break keyword here.

        if index == 0:
            self.insert_at_start(data)
            return

        if index == length:        # insert after last node
            self.insert_at_end(data)
            return
        # For all remaining conditions or to insert in between
        itr = self.head
        count = 0
        while itr:  #itr is node 30
            if index == count:                      #10 20 30 40   -> add 25 after 20 - insert at 2nd index
                node = ll_node(data,itr,itr.prev)
                itr.prev.next = node
                itr.prev = node
                return
            count += 1
            itr = itr.next

# (Exercise:) Define functions for insert after value and remove by value. Add them in the same Doubly_ll class:
# 1) def insert_after_value(self,data_after,data_to_insert):
# 2) def remove_by_value(self,data):

```
